# Intro to numpy

```python
# Creating 1D arrays

arr1d = np.array([1, 2, 3, 4, 5])

print(arr1d)
```

```python
# Creating 2D arrays

arr2d = np.array([[1, 2, 3], [4, 5, 6]])

print(arr2d)
```

```python
# Creating 3D arrays

arr3d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr3d)
```

```python
# Check dimensions of array
print(arr1d.ndim)
print(arr2d.ndim)
print(arr3d.ndim)
```

```python
# Indexing into 1D arrays. Get 3 from arr1d
print(arr1d[2])

# Indexing into 2D arrays. Get from arr2d 4
print(arr2d[1][1])

# Indexing into 3D arrays. Get 2nd 4 from arr 3d
print(arr3d[1][1][0])
```

```python
# Slicing 1D Arrays

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[4:]) # get elements from 4th index to the end of the array
print(arr[:4]) # get elements from the beginning of the array to the 4th index
print(arr[-3:]) # get elements for -3 to end of array
print(arr[:-3]) # get elements from begining of array to -3
print(arr[-3:-1]) # get from -3 (5) to -1 (6)
```
